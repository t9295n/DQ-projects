# Investigating Fandango Movie Ratings

In 2015, Walt Hickey anlyzed Fandango movie ratings and found that their rating system was dishonest. Essentially the ratings that were displayed were rounded up from their actual ratings. The goal of this is to determine if the rating system has changed since Hickey's published findings.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline


data1 = pd.read_csv("fandango_score_comparison.csv")
data2 = pd.read_csv("movie_ratings_16_17.csv")

old_data = data1[['FILM', 'Fandango_Stars', 'Fandango_Ratingvalue', 'Fandango_votes', 'Fandango_Difference']].copy()
new_data = data2[['movie', 'year', 'fandango']].copy()
```

First set of data contains films that have at least 30 fan reviews on Fandango. The second data set contains popular movies "with a significant number of votes." Sampling was not random, so our two sets are not representative for the population we're trying to describe.

Instead, popular movie ratings from before the article was published should be compared with popular movie ratings after the publication. Specifically compare Fandango movie ratings in 2015 and 2016.

Need to extract movies released in 2015 and 2016

```python
old_data.head(2)
```

```python
new_data.head(2)
```

This data set doesn't have a column for year, but we can extract it from the FILM column.

```python
old_data["Year"] = old_data["FILM"].str[-5:-1]

films_2015 = old_data.loc[old_data.Year == "2015"].copy()
films_2016 = new_data.loc[new_data.year == 2016].copy()
```

```python
plt.style.use('fivethirtyeight')

films_2015["Fandango_Stars"].plot.kde(label="2015", legend=True, figsize=(8,6))
films_2016["fandango"].plot.kde(label="2016", legend=True)

plt.xlim((0,5))
plt.xticks(np.arange(0,5.1,0.5))
plt.xlabel("Fandango Rating")
plt.title("Comparing Fandango Ratings, 2015 vs 2016", y=1.1)
```

Both curves are skewed, however the shape of the 2016 curve seems more symmetrical, with an average rating of 4.0, while the 2015 curve is more skewed, with a clear higher average rating. Movies in 2016 in general rated lower than movies in 2015. Maybe the rating system changed, although there's a possibility that moves in 2015 were just better.

# Examining frequency distributions

```python
films_2015["Fandango_Stars"].value_counts(normalize=True).sort_index() * 100
```

```python
films_2016["fandango"].value_counts(normalize=True).sort_index() * 100
```

Looking at the frequency distributions, in 2016 there were much less films given a 4.5 or 5 compared to 2015, instead many more films were given a 4.0 or 3.5. 2016 also had films that were rated 2.5, unlike in 2015. It's unclear about the direction of the rating difference. 

```python
mean_2015 = films_2015["Fandango_Stars"].mean()
median_2015 = films_2015["Fandango_Stars"].median()
mode_2015 = films_2015["Fandango_Stars"].mode()[0]
stats_2015 = [mean_2015, median_2015, mode_2015]

mean_2016 = films_2016["fandango"].mean()
median_2016 = films_2016["fandango"].median()
mode_2016 = films_2016["fandango"].mode()[0]
stats_2016 = [mean_2016, median_2016, mode_2016]

print(stats_2015)
print(stats_2016)
```

Looking at the stats, both years had the same median value, but 2015 had more 4.5 ratings, which led to the higher average rating.

```python
stats = pd.DataFrame()
stats['2015'] = stats_2015
stats['2016'] = stats_2016

stats.index = ['mean', 'median', 'mode']
stats.plot.bar(rot=0, figsize=(8,6))
plt.ylim((0,5))
plt.yticks(np.arange(0,5.1,0.5))
plt.ylabel('Rating')
plt.title('Comparing summary statistics: 2015 vs 2016', y=1.05)
plt.show()
```

On average, popular movies in 2016 were rated lower than popular movies in 2015.
